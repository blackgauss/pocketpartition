```python
from pocketpartition import NumericalSet, Partition, NumericalSemigroup
import random
import pandas as pd
import numpy as np
```

```python
def generate_subset(frobenius_number):
    genus = random.choice(list(range(1, frobenius_number + 1)))
    subset = random.sample(list(range(1, frobenius_number + 1)), genus)
    subset.append(frobenius_number)
    return subset
def random_semigroup(frobenius_number):
    subset = generate_subset(frobenius_number)
    return NumericalSet(subset).atom_monoid()
def small_element(gaps, n):
    idx = 0
    done = False
    for gap in gaps:
        if done:
            break
        if gap + 1 not in gaps:
            idx += 1
            idx2 = 0
            if idx == n:
                return (gap + 1) 
            while (gap + idx2 + 1 not in gaps) and not done:
                if (idx + idx2) == n:
                    value = gap + idx2
                    done = True
                idx2 += 1
    return value if value < max(gaps) else 0
def multiplicity(gaps):
    for gap in gaps:
        if gap + 1 not in gaps:
            return (gap + 1) 
```

```python
small_element({1,2,3,5,9,13}, 3)
```

```python
lst = [random_semigroup(15) for i in range(10000)]
semigroups = [i for n, i in enumerate(lst) if i not in lst[:n]]
semigroups;
```

```python
df = pd.DataFrame(np.transpose(semigroups))
df["Frobenius Number"] = df[0].apply(lambda x: max(x))
df["Multiplicity"] = df[0].apply(lambda x: multiplicity(x))
df["Second Hook"] = df[0].apply(lambda x: max(x) - multiplicity(x))
df["Third Hook"] = df[0].apply(lambda x: int(max(x) - small_element(x,2)) if small_element(x,2) != 0 else None)
df["Genus"] = df[0].apply(len)
df["xGenus"] = (df["Frobenius Number"] + 1) - df["Genus"]
df.sort_values(by="Genus", inplace=True)
```

```python
df.sort_values(by="Third Hook", inplace=True, ascending=False)
df.reset_index(drop=True, inplace=True)
```
